```python
from huggingface_hub import hf_hub_download
import torch

model_path = hf_hub_download(
    repo_id="jane-street/2025-03-10",
    filename="model_3_11.pt",
    repo_type="model")

model = torch.load(
    model_path,
    map_location="cpu",
    weights_only=False)

model.eval()
```

```python
linear_layers = [m for m in model.modules() if isinstance(m, torch.nn.Linear)]
second_last = linear_layers[-2]
```

```python
target_bytes = (-second_last.bias[16:32]).int()

#target_hash = "".join(f"{x:02x}" for x in target_bytes)
target_hash = bytes(int(x) for x in (-second_last.bias[16:32]).int()).hex()

print(target_hash)
```

```python
import hashlib

from wordfreq import top_n_list

target = "c7ef65233c40aa32c2b9ace37595fa7c"

words = [

    w for w in top_n_list("en", 100000)

    if w.isalpha() and w.islower()

]

print(len(words))

target_bytes = bytes.fromhex(target)

for i, a in enumerate(words):

    if i % 1000 == 0:

        print("first word index:", i, a)

    prefix = (a + " ").encode()

    for b in words:

        if hashlib.md5(prefix + b.encode()).digest() == target_bytes:

            print("FOUND:", a, b)

            raise SystemExit
```
